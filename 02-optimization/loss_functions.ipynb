To measure and quantify the discrepancy between a models predictions and the ground truth labels, we use loss functions. They represent how “unhappy” the model is with its current parameters: the larger the loss, the worse the predictions. Since the loss is a function of the model’s parameters, it can be minimized by changing these parameters.

In simple cases, the loss function can have a bowl-shaped appearance, representing a convex function. For convex functions, any local minimum is also guaranteed to be the global minimum, which makes optimization easier. However, when we extend these objective functions to neural networks, the optimization landscape becomes non-convex. Instead of smooth bowl-shaped surfaces, we encounter highly complex and irregular terrains with many local minima, saddle points, and flat regions, making optimization significantly more challenging.

This process belongs to the field of optimization which is concerned with finding the minimum or maximum of functions. In machine learning, optimization focuses on minimizing the loss function in order to improve model performance. There are several approaches for optimizing a model’s parameter weights, each with different strengths and trade-offs.

## 1. Loss&Cost Functions

The loss function measures the error for a single training example by comparing the model’s prediction to the true target value.

The cost function aggregates the loss over a set of examples (such as a mini-batch or the entire training dataset) typically by taking the average (or sometimes the sum) of individual losses.

In many modern contexts, “cost function” and “loss function” are used interchangeably.

## 2. Loss Functions for Regression

### 2.1 Mean Squared Error (MSE)

$$\frac{1}{n} \sum_{i=1}^{n} (\hat{Y}_i - Y_i)^2$$

**+**
- Penalizing Large Errors Strongly
- Always Non-Negative so + and - values do not cancel each other out
- Differentiable and convex curve
- Can be broken down into Variance and Squared Bias
  

**-**
- Sensitive to Outliers
- Different Units with the data (square root if needed)
- Symmetrical Error (Overprediction = Underprediction)

```python
import numpy as np

def mse(y_pred, y_true):
    return np.mean((y_pred - y_true) ** 2)
```

### 2.2 Mean Absolute Error (MAE)

$$\frac{1}{n} \sum_{i=1}^{n} |\hat{Y}_i - Y_i|$$

**+**
- Robust to Outliers
- Interpretable (Same Units)
- Median Prediction
  

**-**
- Not Differnetiable at 0 (subgradient)
- Ignores Severity of Large Errors


```python
import numpy as np

def mae(y_pred, y_true):
    return np.mean(np.abs(y_pred - y_true))
```

### 2.3 Huber Loss
$$

\begin{cases} 
\frac{1}{2}(y - \hat{y})^2 & \text{if } |y - \hat{y}| \le \delta \\
\delta |y - \hat{y}| - \frac{1}{2}\delta^2 & \text{if } |y - \hat{y}| > \delta
\end{cases}
$$

Mixture of MSE and MAE, it is robust to outliers and it maintains smooth convergence. But the delta as a hyperparameter must be manually tuned.




```python
def huber_loss(y_pred, y_true, delta=1.0):
    error = y_true - y_pred
    abs_error = np.abs(error)

    quadratic = 0.5 * error**2
    linear = delta * (abs_error - 0.5 * delta)

    return np.where(abs_error <= delta, quadratic, linear)

def huber_loss_mean(y_pred, y_true, delta=1.0):
    loss = huber_loss_mean(y_pred, y_true, delta)
    return np.mean(loss)
```

## 3. Loss Functions for Classification
### 3.1 Binary Cross Entropy (Log Loss)
$$-\frac{1}{N} \sum_{i=1}^{N} \left[ y_i \log(\hat{y}_i) + (1 - y_i) \log(1 - \hat{y}_i) \right]$$
It is used for a classification model where the output is a probability value between 0 and 1. It heavily penalizes predictions that are confident but wrong.

*In neural networks, BCE is usually paired with a Sigmoid activation function in the final layer*
### 3.2 Categorical Cross-Entropy (CCE)
$$-\frac{1}{N} \sum_{i=1}^{N} \sum_{c=1}^{C} y_{i,c} \log(\hat{y}_{i,c})$$
Multi-class extension of Binary Cross-Entropy. It compares the model's predicted probability distribution across all classes against the true distribution.

*Sparse Categorical Cross-Entropy uses this exact same mathematical formula*
### 3.3 Hinge Loss
It deals not with probabilities but with the raw output scores of a classifier. It forces the model not just to classify correctly, but to classify with a healthy "margin" of confidence.
$$\frac{1}{N} \sum_{i=1}^{N} \max(0, 1 - y_i \cdot \hat{y}_i)$$

```python
import numpy as np
from sklearn.datasets import load_digits

#1. data preprocess
digits = load_digits()
X, y = digits.data, digits.target

#train-test-split
split = int(0.8 * len(X))
X_train, X_test = X[:split], X[split:]
y_train, y_test = y[:split], y[split:]

#distance metrics to compare images
def l1_distance(X, X_train):
    return np.sum(np.abs(X[:, np.newaxis] - X_train[np.newaxis, :]), axis=2)

def l2_distance(X, X_train):
    return np.sqrt(np.sum((X[:, np.newaxis] - X_train[np.newaxis, :]) ** 2, axis=2))

#2. kNN classifier
class KNN():
    def __init__(self, k):
        self.k = k

    def train(self, X, y):
        #knn just memorizes data and there is no actual learning
        self.X = X
        self.y = y

    def predict(self, X, distance="l1"):
        #1-compute distance
        if distance == "l1":
            distances = l1_distance(X, self.X)
        else:
            distances = l2_distance(X, self.X)

        #2-get indices of k nearest neighbors for each test point
        k_nearest_indices = np.argsort(distances, axis=1)[:, :self.k]  

        #3-get their labels
        k_nearest_labels = self.y[k_nearest_indices]  

        #4-majority vote for class
        predictions = np.array([
            np.bincount(row).argmax()
            for row in k_nearest_labels
        ])

        return predictions


#3. training("memorizing")&evaluation
knn = KNN(k=3)
knn.train(X_train, y_train)
preds = knn.predict(X_test, distance="l2")

accuracy = np.mean(preds == y_test)
print(f"Accuracy: {accuracy:.4f}")
```
